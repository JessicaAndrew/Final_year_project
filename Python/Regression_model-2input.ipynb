### Custom Loss Function - Quaternion Distance Metric

```python
class CustomAccuracy(tf.keras.losses.Loss):
  def __init__(self):
    super().__init__()
  def call(self, y_true, y_pred): 
    lamda = tf.math.multiply(y_true, tf.math.l2_normalize(y_pred, axis=-1))
    lamda = tf.reduce_sum(lamda, 1, keepdims=True)
    value = 1 - (lamda)
    value = tf.math.reduce_mean(value)
    return value
```

```python
start_neurons = 16
input_layer = Input((img_size_target, img_size_target, 1))

def build_model2():
    x = Conv2D(start_neurons, (3, 3), activation='relu', padding="same")(input_layer)
    x = tf.keras.layers.BatchNormalization(axis=-1)(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.1)(x)
    
    x = Conv2D(start_neurons*2, (3, 3), activation='relu', padding="same")(x)
    x = tf.keras.layers.BatchNormalization(axis=-1)(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.1)(x)
    
    x = Conv2D(start_neurons*4, (3, 3), activation='relu', padding="same")(x)
    x = tf.keras.layers.BatchNormalization(axis=-1)(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.1)(x)
    
    x = Conv2D(start_neurons*8, (3, 3), activation='relu', padding="same")(x)
    x = tf.keras.layers.BatchNormalization(axis=-1)(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.1)(x)
    
    x = Conv2D(start_neurons*16, (3, 3), activation='relu', padding="same")(x)
    x = tf.keras.layers.BatchNormalization(axis=-1)(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.1)(x)
    
    x = GlobalAveragePooling2D()(x)
    x = Dense(units=128, activation='relu')(x)
    x = Dense(units=4)(x)
    
    model = Model(inputs=[input_layer], outputs=[x])
    return model
```

```python
def customLoss(y_true, y_pred): 
    lamda = tf.math.multiply(y_true, tf.math.l2_normalize(y_pred, axis=-1))
    lamda = tf.reduce_sum(lamda, 1, keepdims=True)
    value = tf.math.acos(lamda)
    return value
```

```python
import math

def quaternion_to_euler(x, y, z, w):
    t0 = +2.0 * (w * x + y * z)
    t1 = +1.0 - 2.0 * (x * x + y * y)
    X = math.degrees(math.atan2(t0, t1))

    t2 = +2.0 * (w * y - z * x)
    t2 = +1.0 if t2 > +1.0 else t2
    t2 = -1.0 if t2 < -1.0 else t2
    Y = math.degrees(math.asin(t2))

    t3 = +2.0 * (w * z + x * y)
    t4 = +1.0 - 2.0 * (y * y + z * z)
    Z = math.degrees(math.atan2(t3, t4))

    return X, Y, Z

x1 = 0
y1 = 0
z1 = 0

x2 = 0
y2 = 0
z2 = 0

x3 = 0
y3 = 0
z3 = 0

for i in range (200):
    x1 = x1 + quaternion_to_euler(value1[i][0], value1[i][1], value1[i][2], value1[i][3])[0]
    y1 = y1 + quaternion_to_euler(value1[i][0], value1[i][1], value1[i][2], value1[i][3])[1]
    z1 = z1 + quaternion_to_euler(value1[i][0], value1[i][1], value1[i][2], value1[i][3])[2]
    
    x2 = x2 + quaternion_to_euler(value2[i][0], value2[i][1], value2[i][2], value2[i][3])[0]
    y2 = y2 + quaternion_to_euler(value2[i][0], value2[i][1], value2[i][2], value2[i][3])[1]
    z2 = z2 + quaternion_to_euler(value2[i][0], value2[i][1], value2[i][2], value2[i][3])[2]
    
    x3 = x3 + quaternion_to_euler(value3[i][0], value3[i][1], value3[i][2], value3[i][3])[0]
    y3 = y3 + quaternion_to_euler(value3[i][0], value3[i][1], value3[i][2], value3[i][3])[1]
    z3 = z3 + quaternion_to_euler(value3[i][0], value3[i][1], value3[i][2], value3[i][3])[2]
    
x1 = x1/200
y1 = y1/200
z1 = z1/200

x2 = x2/200
y2 = y2/200
z2 = z2/200

x3 = x3/200
y3 = y3/200
z3 = z3/200
```
